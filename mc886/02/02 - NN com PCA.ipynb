```python
import numpy as np
import pickle
from sklearn.decomposition import PCA
import sklearn.neural_network as nn
import sklearn.linear_model as lm
import time
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
train = {b'labels' : [], b'data' : []}

for i in range(1, 6):
    aux = unpickle('cifar-10-python/cifar-10-batches-py/data_batch_' + str(i))
    train[b'labels'].extend(aux[b'labels'])
    train[b'data'].extend(aux[b'data'])

train_x = train[b'data']

train_y = train[b'labels']

test = unpickle('cifar-10-python/cifar-10-batches-py/test_batch')


test_x = test[b'data']

test_y = test[b'labels']


pca = PCA(n_components=0.93)
pca.fit(train_x)
train_x = pca.transform(train_x)
test_x = pca.transform(test_x)
```

```python
train = {b'labels' : [], b'data' : []}

for i in range(1, 6):
    aux = unpickle('cifar-10-python/cifar-10-batches-py/data_batch_' + str(i))
    train[b'labels'].extend(aux[b'labels'])
    train[b'data'].extend(aux[b'data'])

train_x = train[b'data']

train_y = train[b'labels']

test = unpickle('cifar-10-python/cifar-10-batches-py/test_batch')


test_x = test[b'data']

test_y = test[b'labels']

mlp = nn.MLPClassifier(hidden_layer_sizes=(2700, ), activation='logistic', solver='lbfgs').fit(train_x, train_y)
pred_y = mlp.predict(test_x)
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
cm = confusion_matrix(test_y, pred_y)
plt.matshow(cm)
plt.show()
```

```python
file = open("DoubleCamadaDePCA.txt", "w")
for y in [600, 1400]:
    for x in [300, 800, 1400, 2000]:
        mlp = nn.MLPClassifier(hidden_layer_sizes=(x, y), activation='relu', solver='adam', verbose=True,
                         learning_rate_init=0.001).fit(train_x, train_y)
        scr = mlp.score(test_x, test_y)
        file.write("MLPClassifier: Neuronios: " + str(x) + " " + str(y) +"  activation: relu  solver: adam  learning_rate_init=0.001 -  score= " + str(scr) + "\n")

        mlp = nn.MLPClassifier(hidden_layer_sizes=(x, y), activation='logistic', solver='adam', verbose=True,
                            learning_rate_init=0.001).fit(train_x, train_y)
        scr = mlp.score(test_x, test_y)
        file.write("MLPClassifier: Neuronios: " + str(x) + " " + str(y) +"  activation: logistic  solver: adam  learning_rate_init=0.001 -  score= " + str(scr) + "\n")

        mlp = nn.MLPClassifier(hidden_layer_sizes=(x, y), activation='relu', solver='lbfgs', verbose=True,
                            learning_rate_init=0.001).fit(train_x, train_y)
        scr = mlp.score(test_x, test_y)
        file.write("MLPClassifier: Neuronios: " + str(x) + " " + str(y) +"  activation: relu  solver: lbfgs  learning_rate_init=0.001 -  score= " + str(scr) + "\n")

        mlp = nn.MLPClassifier(hidden_layer_sizes=(x, y), activation='logistic', solver='lbfgs', verbose=True,
                            learning_rate_init=0.001).fit(train_x, train_y)
        scr = mlp.score(test_x, test_y)
        file.write("MLPClassifier: Neuronios: " + str(x) + " " + str(y) +"  activation: logistic  solver: lbfgs  learning_rate_init=0.001 -  score= " + str(scr) + "\n")
        
        file.flush()
        print("I'm alive !!\n")

file.close()
```
